### Часть 5. (1.5 баллов) CVXPY

Для анализа изображений часто используется метод неотрицательной матричной факторизации ([Nonnegative matrix factorization](https://arxiv.org/pdf/1401.5226.pdf)).
Идея в том, чтобы решить следующую задачу

\begin{align*}
& \min_{W, H} \frac12 \|X - WH\|^2_F \\
\text{s.t. } & W \geq 0\\
& H \geq 0,
\end{align*}

где знак $\geq$ означает покомпонентное сравнение.
Также матрицы $W$ и $H$ малого ранга, то есть $W \in \mathbb{R}^{m \times k}$ и $H \in \mathbb{R}^{k \times n}$, где $k$ - параметр, который необходимо настраивать.
Чем больше $k$, тем точнее приближение, но и тем дольше будет решаться задача.
Матрица $X \in \mathbb{R}^{m \times n}$, где $m$ - это число фото в нашем наборе, а $n$ - это количество пикселей в каждой картинке, то есть каждая строка матрицы $X$ - это векторизованная картинка.

**Задание 0. (0.15 баллов)** Как можно интерпретировать факторы $W$ и $H$ и почему важна их покомпонентная неотрицательность?

**Ответ:**
Факторы $W$ и $H$ в методе неотрицательной матричной факторизации (NMF) имеют следующие интерпретации:

Матрица $W$: это матрица базисных изображений или компонент. Каждый столбец матрицы $W$ представляет собой базисное изображение, которое может быть интерпретировано как характеристический образец или паттерн, присутствующий в данных. Например, если NMF применяется к коллекции изображений лиц, каждый столбец $W$ может представлять собой фрагменты или черты лиц, такие как глаза, нос, рот и т.д.

Матрица $H$: это матрица весов или коэффициентов, которая определяет, как каждое изображение в данных может быть представлено в терминах компонентов. Каждый столбец матрицы $H$ представляет собой вектор весов, показывающий вклад каждой компоненты в соответствующее изображение. Например, если столбец $H$ имеет большие значения в первых нескольких компонентах, это может означать, что соответствующее изображение содержит много черт лица, представленных этими компонентами.

Покомпонентная неотрицательность является ключевым аспектом NMF и важна по следующим причинам:

Интерпретируемость: положительность компонентов в матрице $W$ и коэффициентов в матрице $H$ позволяет интерпретировать каждую компоненту как положительный шаблон или признак, который присутствует в данных. Это делает результаты NMF более понятными и интерпретируемыми для человека.

Естественность: в многих приложениях данные не могут быть отрицательными, например, интенсивности пикселей изображений, звуковые сигналы или концентрации веществ в химических образцах. Поэтому требование неотрицательности позволяет более естественно моделировать такие данные.


Такую задачу можно решать с помощью cvxpy, если воспользоваться идеей попеременной оптимизации (alternating optimization). Идея состоит в том, чтобы на чётных итерациях решить точно задачу относительно параметра $W$ при фиксированном параметре $H$, а на нечётных наоборот: использовать решение для $W$ с предыдущей итерации в качестве фиксированного параметра и найти решение задачи относительно $H$ и так далее до сходимости.
Сначала Вам нужно инициализировать матрицы $W$ и $H$ какими-нибудь матрицами с неотрицательными элементами.

Загрузим данные. Мы будем раскладывать изображения лиц. О таком подходе к анализу фотографий лиц вам рассказывали или расскажут в курсе ОиРИ, а тут вы просто реализуете этот метод

````python
import cvxpy as cvx
import sklearn.datasets as skldata
import matplotlib.pyplot as plt
import sklearn.decomposition as skdec
%matplotlib inline
data = skldata.fetch_olivetti_faces()
import numpy as np
print("Shape of data = {}".format(data["images"].shape))
index = 42
plt.imshow(data["images"][index], cmap=plt.cm.Greys)
plt.colorbar()
_ = plt.axis("off")
````

**Задание 1. (1.05 балл)** Реализуйте с помощью cvxpy идею попеременной оптимизации для этой задачи и решите её

*Совет от Соника:* Сперва тестируйте алгоритм не на всех данных, так как вычисления занимают большое количество времени

*Замечание*: все еще можно менять интерфейсы функций

````python
def non_negative_matrix_factorization(X, rank, max_iters, tol, log=False):
    """
    X - матрица изображений. Каждое изображение является строкой
    rank - какой будете использовать ранг
    max_iters - максимальное количество итераций
    tol - точность, с какой решается задача
    log - флаг, отвечающий за то, будете ли вы печатать логи или нет (такое желание может возникнуть, так как
    алгоритм будет работать долго)
    Функция возвращает найденные матрицы H и W, а также значения функционала в ходе оптимизации
    """
    objective_values = []
    H = np.random.rand(rank, X.shape[1])
    W = np.random.rand(X.shape[0], rank)
    objective_values.append(0.5 * np.linalg.norm(X - W @ H, 'fro') ** 2)
    for iter in range(max_iters):
      if iter % 2 == 0:
        W_variable = cvx.Variable((X.shape[0], rank), nonneg=True)
        W_problem = cvx.Problem(cvx.Minimize(0.5 * cvx.norm(X - W_variable @ H, 'fro') ** 2), [])
        W_problem.solve(solver=cvx.SCS)
        W = W_variable.value
      else:
        H_variable = cvx.Variable((rank, X.shape[1]), nonneg=True)
        H_problem = cvx.Problem(cvx.Minimize(0.5 * cvx.norm(X - W @ H_variable, 'fro') ** 2), [])
        H_problem.solve(solver=cvx.SCS)
        H = H_variable.value
      objective_values.append(0.5 * np.linalg.norm(X - W @ H, 'fro') ** 2)
      if log:
        print(f'Iter: {iter}; Objective value: {objective_values[-1]}')
      if np.abs(objective_values[-1] - objective_values[-2]) < tol:
        break
    return H, W, objective_values
````

````python
np.random.seed(0)
X = data["images"].reshape(data["images"].shape[0], -1)
H, W, objective_values = non_negative_matrix_factorization(X, 10, 20, 1e-1, True)
````
